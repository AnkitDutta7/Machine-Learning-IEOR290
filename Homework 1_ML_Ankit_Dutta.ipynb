```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
graph= plt.figure()
ax=graph.add_subplot(111)

a=np.linspace(-1,1,100)
b=np.linspace(-0.5,0.5,100)

x,y=np.meshgrid(a,b)
z=(x-y**2)*(x-2*y**2)

cplt=plt.contour(x,y,z)
plt.clabel(cplt,inline=True)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import numpy.random


a=np.linspace(-1,1,100)
b=np.linspace(-0.5,0.5,100)

x,y=np.meshgrid(a,b)
z=(x-y**2)*(x-2*y**2)
plt.pcolormesh(z)
plt.colorbar()

cplt=plt.contour(z, colors="white")
plt.clabel(cplt,inline=True)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
