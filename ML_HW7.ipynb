```python
from mnist import MNIST
import numpy as np
import matplotlib.pyplot as plt
import sklearn.metrics as metrics
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import OneHotEncoder
```

```python
def load_dataset():
    mndata = MNIST('./data/')
    X_train, labels_train = map(np.array, mndata.load_training())
    X_test, labels_test = map(np.array, mndata.load_testing())
    # Scale x
    X_train = X_train/255.0
    X_test = X_test/255.0
    labels_train = np.asarray(labels_train).reshape(60000,1) #OneHotEncoder only accepts 2D
    labels_test = np.asarray(labels_test).reshape(10000,1) #OneHotEncoder only accepts 2D
    return X_train, labels_train, X_test, labels_test
```

```python
def plot(x):
    image = x.reshape(28,28)
    plt.imshow(image, cmap='winter')
    plt.axis('off')
    plt.show()
```

```python
# To run this in ipython, use command %run <filename>
if __name__ == "__main__":
# Read data
    X_train, labels_train, X_test, labels_test = load_dataset()
    
    # One hot encode
    enc = OneHotEncoder()
    enc.fit (labels_train) # Find max/min categorical values for each column
    Y_train = enc.transform(labels_train).toarray()
    
    # Training using Linear Regression
    mlp = MLPClassifier(hidden_layer_sizes=(200,),
                        activation='relu',
                        solver='lbfgs', 
                        alpha=1e-5,
                        random_state=2,
                        early_stopping=False)
    mlp.fit(X_train, Y_train)
    
    # Predict
    Pred_Y_train = mlp.predict(X_train)
    Pred_Y_test = mlp.predict(X_test)
    
    # Decode the value
    pred_labels_train = np.argmax(Pred_Y_train, axis=1)
    pred_labels_train = np.asarray(pred_labels_train).reshape(60000,1) #OneHotEncoder only accepts 2D
    pred_labels_test = np.argmax(Pred_Y_test, axis=1)
    pred_labels_test = np.asarray(pred_labels_test).reshape(10000,1) #OneHotEncoder only accepts 2D
    
    # Print accuracy
    print("Train accuracy: {0}".format(metrics.accuracy_score(labels_train, pred_labels_train)))
    print("Test accuracy: {0}".format(metrics.accuracy_score(labels_test, pred_labels_test)))
    
    
    # Plot first mis-classified data
    delta = pred_labels_test-labels_test
    first_wrong = np.where(delta!=0)[0][0]
    print("The first image that gets wrong by our model:")
    plot(X_test[first_wrong])
    print("Predicted value by model: {0}".format(pred_labels_test[first_wrong][0]))
```
